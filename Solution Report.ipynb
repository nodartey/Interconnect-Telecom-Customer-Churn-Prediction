## Interconnect Telecom Customer Churn Prediction

**Introduction**

The goal of this project was to develop a machine learning model for the telecom operator Interconnect to predict whether a customer will churn (i.e., end their contract). The company provided several datasets related to customer demographics, contract details, internet services, and phone services.

To ensure customer retention and reduce revenue loss, it is important to accurately identify customers at risk of churning. A reliable predictive model would allow the company to proactively target these customers with personalized offers and support.

The main evaluation metric for this task was **AUC-ROC**, as it provides a robust measure of classification performance, especially for imbalanced datasets. Secondary metrics such as **accuracy**, **precision**, **recall**, and **F1-score** were also used to evaluate the quality of predictions.

This report summarizes all the steps taken during the solution process, the challenges encountered, and the rationale behind the final model selection.

### Question 3: What is your final model and what quality score does it have?

**The final model selected is Logistic Regression, chosen because it produced the highest AUC-ROC score (0.8279) among all models, while also achieving strong performance across other classification metrics.**

It achieved the following metrics and it's impact on Interconnect Telecommunication:

1. **AUC-ROC: 0.8279**

What it means:

- AUC-ROC (Area Under the Receiver Operating Characteristic curve) measures how well the model distinguishes between churners and non-churners.

- A value close to 1 means the model is very good at distinguishing between the two classes.

**Business impact:**

- A high AUC-ROC of 0.8279 means the model can confidently prioritize customers by their likelihood to churn.

- This allows the business to efficiently allocate retention resources (e.g., discounts, outreach) to those at highest risk.


2. **Accuracy: 0.7330**

What it means:

- Accuracy measures how many predictions the model got right out of all predictions.

**Business impact:**

- About 73% of customers were correctly classified as either churning or staying.

- However, since churn is often imbalanced (fewer churners), this metric can be misleading. That’s why recall and precision for the churn class are more useful in this case.


3. **Recall (Churn class): 0.8042**

What it means:

- Recall is the proportion of actual churners that the model correctly identified.

- In other words, out of all customers who actually churned, 80.4% were correctly flagged.

**Business impact:**

- This is critical for churn prevention: the company will successfully identify most customers at risk of leaving.

- It means proactive intervention strategies (loyalty offers, personalized plans) can be applied before it's too late.

  

4. **Precision (Churn class): 0.4991**

What it means:

- Precision is the proportion of predicted churners who actually did churn.

- So, about 50% of customers flagged as churners truly are.

**Business impact:**

- There's a trade-off here: some resources may be used on customers who wouldn’t have churned.

- However, because retaining a customer is usually cheaper than acquiring a new one, this is often acceptable.

- The business might fine-tune interventions (e.g., lower-cost incentives) to manage cost while still reducing churn.



5. **F1 Score: 0.6159**

What it means:

- The F1 score balances precision and recall: it gives a single score that reflects both false positives and false negatives.

- This is especially useful when the data is imbalanced (like churn).

**Business impact:**

- An F1 score of 61.6% reflects a solid balance: the model captures most churners without overwhelming the company with too many false alarms.

- It indicates the model is suitable for a real-world deployment scenario.



###  Business Recommendations for Stakeholders

**Based on the model’s insights, we recommend the following actions for Interconnect’s marketing and retention teams:**

1. **Prioritize High-Risk Customers Identified by the Model**
- **Why:** The model achieves 80.4% recall, meaning it can detect most customers who are likely to churn.

- **Recommendation:** Develop a churn prevention program targeting high-risk customers with proactive offers such as:

   - Loyalty discounts

   - Personalized data plans

   - Free add-ons (e.g., streaming subscriptions, bonus data)

- Focus on early intervention before customers decide to switch providers.


2. **Optimize Retention Costs Using Precision Insights**
- **Why:** Precision is around 49.9%, meaning about half of the flagged churners truly intend to leave.

**Recommendation:**

- Use a **tiered retention strategy:**

     - High-confidence churners → High-value offers

     - Lower-confidence churners → Low-cost touchpoints (SMS reminders, satisfaction surveys)

- This avoids overspending on customers who were not at risk.


3. **Integrate Model Insights into CRM and Call Center Tools**
- **Why:** Real-time access to churn scores can help frontline staff tailor interactions.

- **Recommendation:**

   - Embed model predictions into CRM dashboards.

   - Train call center agents to offer **custom solutions** based on churn risk level (e.g., faster escalation, proactive upgrades).

   - Include customer churn score as a flag during each service interaction.


4. **Enhance Customer Experience Based on Feature Importance**
- **Why:** Churn models often rely on behavioral and contract-based features (e.g., contract length, service usage, tech support calls).

- **Recommendation:**

   - Use SHAP or feature importance analysis (if not done already) to understand why customers churn.

   - Improve **onboarding, billing transparency, and network reliability** in segments with high churn rates.

   - Simplify plan migration to reduce customer frustration and improve satisfaction.


5. **Monitor and Retrain the Model Regularly**
- **Why:** Customer behavior and market competition can change over time.

- **Recommendation:**

    - **Retrain the model quarterly** using fresh data to maintain accuracy.

    - Monitor **drift in model performance** (AUC, recall) and churn patterns.

    - Evaluate model performance separately for **new customers vs. long-term customers.**


6. **Segment Churn Prevention by Customer Lifetime Value (CLV)**
- **Why:** Not all churns are equally costly.

- **Recommendation:**

   - Combine churn risk with **CLV scoring.**

   - Retain high-risk, high-value customers aggressively.

   - Let go of low-risk, low-value churners or offer only low-cost interventions.
 

7. **Launch a "Win-Back" Campaign**
- **Why:** Even missed churners (20% the model didn’t catch) represent recoverable value.

**Recommendation:**

- Use past churner lists and personalize **email/SMS win-back campaigns.**

- Offer limited-time promotions and ask for feedback on why they left.


8. **Measure ROI of Retention Campaigns**
- **Why:** Stakeholders need to track effectiveness of model-guided actions.

- **Recommendation:**

   - Set up **A/B testing** to compare retention outcomes with and without model guidance.

   - Track churn rate, retention cost, and revenue impact over time.
 

**Summary for Stakeholders:**

Interconnect Telecom can significantly reduce churn by using the model to target at-risk customers early, allocate retention budgets wisely, and embed churn insights into business operations. A structured churn mitigation plan — backed by predictive analytics — will improve customer loyalty, protect revenue, and enhance competitive advantage in the telecom market.
